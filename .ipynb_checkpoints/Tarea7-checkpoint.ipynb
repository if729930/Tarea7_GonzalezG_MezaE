Aproxime el valor de la siguiente integral usando el método monte carlo crudo y método de reducción de varianza de muestreo estratíficado

$$I=\int_{0}^{1}x^2\text{d}x=\left.\frac{x^3}{3}\right|_{x=0}^{x=1}=\frac{1}{3}\approx 0.33333$$

Pasos

```python
# Integración montecarlo
def int_montecarlo(f:'Función a integrar',
                   a:'Límite inferior de la integral',
                   b:'Límite superior de la integral',
                   U:'Muestra de números U~[a,b]'):
    return (b-a)/len(U)*np.sum(f(U))
            
    
def estrat(N):
    
    return (np.random.rand(N) + np.arange(0, N))/ N
```

```python
I = 1/3

#Intervalo de integración
a,b = 0,1

#Cantidad de términos 
N = np.logspace(1,7,7, dtype = int)


#Tabla
df = pd.DataFrame(index=N,columns=['Montecarlo crudo', 'Error_relativo1%', "Muestreo estratificado", "Error_relativo2%"], dtype='float')
df.index.name = "Cant_terminos"


#Variables aleatorias en el intervalo [a,b]
ui = list(map(lambda n: np.random.uniform(a,b,n),N))
#Variables muestreo estratificado
ui_estrat = list(map(lambda n: estrat(n),N))


#Aproximación
I_mont = list(map(lambda Y:int_montecarlo(lambda x:x**2,a,b,Y),ui))
I_estrat = list(map(lambda Y:int_montecarlo(lambda x:x**2,a,b,Y),ui_estrat))

#Resultados en la tabla
df.loc[N,"Montecarlo crudo"] = I_mont
df.loc[N, "Muestreo estratificado"] = I_estrat
df.loc[N,"Error_relativo1%"] = np.abs(df.loc[N,"Montecarlo crudo"]-I)*100/I
df.loc[N,"Error_relativo2%"] = np.abs(df.loc[N,"Muestreo estratificado"]-I)*100/I
df
```

```python
def int_montecarlo(f:'Función a integrar',
                   a:'Límite inferior de la integral',
                   b:'Límite superior de la integral',
                   U:'Muestra de números U~[a,b]'):
    return (b-a)/len(U)*np.sum(f(U))

#  Muestreo estratificado 
def Muestreo_estratificado(B): #B son los # de estratos
    U2=np.random.rand(B)
    i=np.arange(0,B)
    v=(U2+i)/B
    return v
```

```python
np.random.seed(5555)
f=lambda x:x**2
f_inv=lambda u: (3*u)**I
```

```python
I = 1/3 #valor real para compararlo
# intervalo de integración
a = 0; b = 1 

# Cantidad de términos, en escala logarítmica
N = np.logspace(1,7,7,dtype=int) # intervalo de 1-7, 7 valores
#se pone en escala logaritmica porque queremos ir aumentando 
# es lo de 10,100,1000,1000....

# Definimos la tabla donde se mostrarán los resultados
df = pd.DataFrame(index=N,columns=['Montecarlo crudo', 'Error_relativo1%'], dtype='float')
df.index.name = "Cantidad_terminos"

# Números aleatorios dependiente de la cantidad de términos N
ui = list(map(lambda N:np.random.uniform(a,b,N),N)) #numeros uniformes 
ui_estra=list(map(lambda x: Muestreo_estratificado(x),N))

# Calculamos la aproximación por montecarlo dependiendo de la cantidad de
# términos que hayamos creado con ui
I_m = list(map(lambda Y:int_montecarlo(lambda x:x**2,a,b,Y),ui)) #funcion, parametros de entrada ,Y, llamamos la funcion con la lista ui
muestras= list(map(lambda x : f(x),ui_estra))
medias = (list(map(lambda x : np.mean(x),muestras)))


# Mostramos los resultados en la tabla previamente creada
df.loc[N,"Montecarlo crudo"] = I_m
df.loc[N,"Error_relativo1%"] = np.abs(df.loc[N,"Montecarlo crudo"]-I)*100/I
df.loc[N,'Muestreo estratificado']=medias
df.loc[N,'Error_relativo2%']=np.abs(df.loc[N,'Muestreo estratificado']-(I))*100/(I)
df
```
